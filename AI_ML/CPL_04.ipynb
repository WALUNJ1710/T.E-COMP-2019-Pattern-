# PERFORM  BASIC PCA ON THE IRIS DATASET. 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("/content/iris.csv")
iris.head()
```

```python
iris.describe()
```

## Correlation matrix

```python
corr = iris.corr(method ="pearson")
display(corr)
sns.heatmap(corr,cmap =sns.light_palette('b', n_colors=6, reverse=False, as_cmap=False, input='rgb'),annot= True,fmt = "f")


```

    Petal_width <-> Sepal_Length : 0.817941 (Positive Correlation)

## PCA (Principal Component Analysis)

```python
from sklearn.decomposition import PCA
pca = PCA()
x_new1 =pca.fit_transform(iris.drop(["Species"],axis =1))
x_new1[:5]
```

```python
explained_variance= pca.explained_variance_ratio_

explained_variance
```

```python
plt.figure(figsize =(6,4))
plt.bar(range(4),explained_variance, alpha=0.5, align='center', label ="Individual Explained Variance", color ="g")
plt.ylabel("Explained Variance Ratio")
plt.xlabel("Principal Component")
plt.legend(loc ="best")
plt.tight_layout()
```
